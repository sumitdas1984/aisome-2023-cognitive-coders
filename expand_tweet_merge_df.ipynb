```python
import pandas as pd
```

```python
country_df = pd.read_csv("country_df.csv")
political_df = pd.read_csv("political_df.csv")
religious_df1 = pd.read_csv("religious_df.csv")
religious_df = religious_df1[religious_df1['labels'] == 'religious']
conspiracy_df = pd.read_csv("conspiracy_df.csv")
concatenated_df = pd.concat([country_df, political_df, religious_df, conspiracy_df], ignore_index=True)
```

```python
concatenated_df.drop(columns=['with_entities'], inplace=True)
concatenated_df.to_csv("tweet_extended.csv")
```
